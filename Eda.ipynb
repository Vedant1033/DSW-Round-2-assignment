```python
# Data Manipulation
import pandas as pd
import numpy as np

# Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Statistical Analysis
from scipy.stats import skew, kurtosis
```

```python
# Load train data
train_data_path = 'train_data.xlsx'
train_data = pd.read_excel(train_data_path)

# Summary Statistics
summary_stats = train_data.describe(include='all').transpose()
```

```python

# Missing Values
missing_values = train_data.isnull().sum()

# Skewness and Kurtosis for numerical features
numerical_cols = train_data.select_dtypes(include=['float64', 'int64']).columns
skewness = train_data[numerical_cols].apply(lambda x: skew(x.dropna()))
kurtosis_values = train_data[numerical_cols].apply(lambda x: kurtosis(x.dropna()))
```

```python
# Correlation Matrix
correlation_matrix = train_data.select_dtypes(include=np.number).corr()

# Visualizations
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
```

```python
# Distribution plots for numerical columns
dist_plots = {}
for col in numerical_cols:
    plt.figure()
    sns.histplot(train_data[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    dist_plots[col] = plt.gcf()
```

```python
# Boxplots for numerical features to detect outliers
for col in numerical_cols:
    plt.figure()
    sns.boxplot(x=train_data[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()
```

```python
# Categorical Columns Analysis
categorical_cols = train_data.select_dtypes(include=['object']).columns
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(y=col, data=train_data, order=train_data[col].value_counts().index)
    plt.title(f'Count Plot for {col}')
    plt.tight_layout()
```

```python
# Output EDA results
summary_stats = train_data.describe(include='all').transpose()  # Calculate summary stats
```

```python
# Print EDA results
print("## SUMMARY STATISTICS:")
print(summary_stats)

print("\n## MISSING VALUES:")
print(missing_values)

print("\n## SKEWNESS:")
print(skewness)

print("\n## KURTOSIS:")
print(kurtosis_values)

```
